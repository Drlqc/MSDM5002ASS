```python
import numpy as np
import matplotlib.pyplot as plt

T=np.linspace(0,2*np.pi,1024)
x=(1-np.cos(T))*np.cos(T)
y=(1-np.cos(T))*np.sin(T)
plt.subplot(221)
plt.plot(y,x,linewidth=3.0,color='b')
plt.title("20787264")

plt.subplot(222)
x1=np.linspace(-2,2,1024)
y1=np.sqrt(1-(abs(x1)-1)*(abs(x1)-1))
y2=np.arccos(1-abs(x1))-np.pi
plt.plot(x1,y1,linewidth=3.0,color='r')
plt.plot(x1,y2,linewidth=3.0,color='r')
plt.title("Li,Qichao")

plt.subplot(223)
x2=np.linspace(-2,2,1024)
y3=np.sqrt(1-(abs(x1)-1)*(abs(x1)-1))
y4=np.arccos(1-abs(x1))-np.pi
plt.plot(x2,y3,linewidth=3.0,color='r')
plt.plot(x2,y4,linewidth=3.0,color='r')
plt.text(-0.5,-1.0,"Li,Qichao",color='b',fontsize=10)

plt.axis('off')


#plt.figure(figsize=(10,10),dpi=100)
plt.subplot(224,polar=True)
#plt.figure(figsize=(10,10),dpi=100)
plt.plot(T,1-np.sin(T),color='b')

plt.show()
```
